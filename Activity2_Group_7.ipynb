### Understanding `for` Loops

A `for` loop is used for iterating over a sequence. Here is the basic syntax:

```python
for item in sequence:
    # execute code
```

The `range()` function is commonly used to loop a specific number of times.

````python
# Example 1: Iterating through a list of fruits
fruits = ["apple", "banana", "cherry", "cherry", "watermelon", "tomato"]
print("Printing fruits:")
for fruit in fruits:
    print(f"Certainly a fruit: {fruit}")

# Example 2: Using range to count to 5
print("\nCounting numbers:")
for i in range(1, 11):
    print(i)
````

Small project idea: Grade calculator.

````python
# Grade Calculator - Version 1

try:
    # Get user input
    score_input = input("Enter the score (0-100): ")
    score = float(score_input)

    # Determine Grade and Status
    if 90 <= score <= 100:
        grade = "A"
        status = "Pass"
    elif 80 <= score < 90:
        grade = "B"
        status = "Pass"
    elif 70 <= score < 80:
        grade = "C"
        status = "Pass"
    elif 60 <= score < 70:
        grade = "D"
        status = "Pass"
    elif 0 <= score < 60:
        grade = "F"
        status = "Fail"
    else:
        grade = "N/A"
        status = "Error: Score out of range (0-100)"

    # Print results
    print(f"Letter Grade: {grade}")
    print(f"Status: {status}")

except ValueError:
    print("Error: Please enter a valid numerical value.")
````

First we will transform this into a more Finland friendly version with grades from 1-5.

Changes we will implement: change the grading system, make the program into a function, change if-else statements into a continuous while true loop, make it more user friendly.

````python
def classify_grade(score):

    if score >= 90: return "5 (Excellent!)"
    if score >= 80: return "4 (Great job!)"
    if score >= 70: return "3 (Good)"
    if score >= 60: return "2 (Passed)"
    if score >= 50: return "1 (Passed)"
    return "F (Fail)"



while True:
    user_input = input("\nEnter score (0-100) or 'q' to quit: ")

    if user_input.lower() == 'q':
        print("Goodbye!")
        break

    try:
        score_val = float(user_input)
        if 0 <= score_val <= 100:
            print(f"Result: {classify_grade(score_val)}")
        else:
            print("Please enter a number between 0 and 100.")
    except ValueError:
        print("Invalid input. Please enter a numerical score.")
````

Only hard thing was the standardization of input for breaking out of the loop (q).
